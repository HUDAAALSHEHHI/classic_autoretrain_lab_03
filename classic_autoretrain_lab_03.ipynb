```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.linear_model import LogisticRegression
import numpy as np
import joblib

# initial dataset
data = {
    "text": ["good", "bad", "excellent", "terrible", "happy", "sad"],
    "label": [1,0,1,0,1,0]
}
df = pd.DataFrame(data)

X_train, X_test, y_train, y_test = train_test_split(df["text"], df["label"], test_size=0.2)
vec = CountVectorizer()
X_train_vec = vec.fit_transform(X_train)
clf = LogisticRegression()
clf.fit(X_train_vec, y_train)

joblib.dump((clf, vec), "model.joblib")
```

```python
new_data = pd.Series(["awful", "boring", "horrible", "bad", "poor"])
X_new = vec.transform(new_data)
new_preds = clf.predict(X_new)

baseline_mean = np.mean(y_train)
new_mean = np.mean(new_preds)
diff = abs(baseline_mean - new_mean)
diff
```

```python
THRESHOLD = 0.2  # trigger condition

if diff > THRESHOLD:
    df2 = pd.DataFrame({"text": new_data, "label":[0,0,0,0,0]})
    df_final = pd.concat([df, df2])

    X_train2, _, y_train2, _ = train_test_split(df_final["text"], df_final["label"], test_size=0.2)
    X_train2_vec = vec.fit_transform(X_train2)
    clf.fit(X_train2_vec, y_train2)
    joblib.dump((clf, vec), "model.joblib")
    print("Model retrained due to performance drop.")
else:
    print("No retraining required.")
```
